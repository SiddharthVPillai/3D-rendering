```python
import os
from PIL import Image

def images_to_gif(input_dir, output_path='output.gif', duration=100, loop=0):
    # Get all PNG files in the input directory
    png_files = [f for f in os.listdir(input_dir) if f.endswith('.png')]

    # Sort files to maintain order
    png_files.sort()

    # Open the first image to get the mode and size
    images = []
    for png_file in png_files:
        img_path = os.path.join(input_dir, png_file)
        img = Image.open(img_path)
        images.append(img.convert('RGB'))

    # Save as GIF
    images[0].save(
        output_path,
        save_all=True,
        append_images=images[1:],
        duration=duration,  # duration between frames in milliseconds
        loop=loop  # 0 means loop indefinitely
    )

    print(f'GIF created: {output_path}')
    print(f'Total frames: {len(images)}')

# Example usage
input_directory = '/content/gaussian-splatting/output/b552c6b4-a/train/ours_25000/renders'
output_gif_path = '/content/gaussian-splatting/output/b552c6b4-a/49_img_25k.gif'

images_to_gif(input_directory, output_path=output_gif_path)
```

```python
import os
import cv2
import numpy as np

def images_to_video(input_dir, output_path='output.mp4', fps=30):
    # Get all PNG files in the input directory
    png_files = [f for f in os.listdir(input_dir) if f.endswith('.png')]

    # Sort files to maintain order
    png_files.sort()

    # Read the first image to get video dimensions
    first_image_path = os.path.join(input_dir, png_files[0])
    frame = cv2.imread(first_image_path)
    height, width, layers = frame.shape

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    video = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # Write frames to video
    for png_file in png_files:
        img_path = os.path.join(input_dir, png_file)
        frame = cv2.imread(img_path)
        video.write(frame)

    # Release the video writer
    video.release()

    print(f'Video created: {output_path}')
    print(f'Total frames: {len(png_files)}')
    print(f'Video dimensions: {width}x{height}')
    print(f'Frame rate: {fps} fps')

# Example usage
input_directory = '/content/gaussian-splatting/output/b552c6b4-a/train/ours_25000/renders'
output_video_path = '/content/gaussian-splatting/output/b552c6b4-a/49_img_25k.mp4'

images_to_video(input_directory, output_path=output_video_path)
```
